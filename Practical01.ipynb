## Basics of Statistical Machine Lerning

Mean, Mode and Median of Data  
Standard Deviation and Variance

```python
# Mean, Mode and Median of Data

import numpy as np
from scipy import stats

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x = np.mean(speed)
y = np.median(speed)

print(x)
print(y)

a = stats.mode(speed)
print(a)

# standard deviation
z = np.std(speed)
print(z)

# calculate variance
v = np.var(speed)
print(v)
```

```python
# Percentiles
ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]

# What is the age that 75% of the people are younger than?
per = np.percentile(ages, 75)
print(per)

# What is the age that 90% of the people are younger than?
x = np.percentile(ages, 90)

print(x)
```

Histogram Explained  

We use the array from the example above to draw a histogram with 5 bars.

The first bar represents how many values in the array are between 0 and 1.

The second bar represents how many values are between 1 and 2. & so on.

```python
# Draw Histogram

import matplotlib.pyplot as plt

x = np.random.uniform(0.0, 5.0, 250)

plt.hist(x, 5)
plt.show()
```

```python
# Create an array with 100000 random numbers, and display them using a histogram with 100 bars

x = np.random.uniform(0.0, 5.0, 100000)

plt.hist(x, 100)
plt.show()
```

### Linear Regression

In the example below, the x-axis represents age, and the y-axis represents speed. We have registered the age and speed of 13 cars as they were passing a tollbooth. Let us see if the data we collected could be used in a linear regression

```python
# Linear Regression

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

```python
slope, intercept, r, p, std_err = stats.linregress(x, y)

print(slope, intercept, r, p, std_err)
```

```python
# Predict the future value

speed = myfunc(10)

print(speed)

s1 = myfunc(15)
print(s1)
```

```python
x = [89,43,36,36,95,10,66,34,38,20,26,29,48,64,6,5,36,66,72,40]
y = [21,46,3,35,67,95,53,72,58,10,26,34,90,33,38,20,56,2,47,15]

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

```python
# Polynomial Regression

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

plt.scatter(x, y)
plt.show()
```

```python
from sklearn.metrics import r2_score

mymodel = np.poly1d(np.polyfit(x, y, 3))

myline = np.linspace(1, 22, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```

```python
print(r2_score(y, mymodel(x)))
```

```python
# Predict future values


mymodel = np.poly1d(np.polyfit(x, y, 3))

speed = mymodel(17)
print(speed)
```

```python
# Let us create an example where polynomial regression would not be the best method to predict future values.

x = [89,43,36,36,95,10,66,34,38,20,26,29,48,64,6,5,36,66,72,40]
y = [21,46,3,35,67,95,53,72,58,10,26,34,90,33,38,20,56,2,47,15]

mymodel = np.poly1d(np.polyfit(x, y, 3))

myline = np.linspace(2, 95, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```

```python
print(r2_score(y, mymodel(x)))
```
